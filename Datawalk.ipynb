```python
import json

# read file
with open('ex2.metadata.json', 'r') as myfile:
    data=myfile.read()

# parse file
json_object = json.loads(data)
```

```python

json_formatted_str = json.dumps(json_object, indent=2)

print(json_formatted_str)
```

```python
new_json={}
new_json["classes_with_virtual_attribute"]=[]
```

```python
classes = json_object["classes"]
for key, value in classes.items():
    new_record={}
    for k, v in value.items():
        if k=="name":
            new_record["name"]=v
        if k=="attributes":
            new_record["attributes"]=[]
            for vv in v:
                if  vv['virtual']==True:
                    new_rr={}
                    new_rr["name"]=vv["name"]
                    new_rr["id"]=vv["id"]
                    new_record["attributes"].append(new_rr)
    if len(new_record["attributes"])>0:            
        new_json["classes_with_virtual_attribute"].append(new_record)            
```

```python
 print(json.dumps(new_json, indent=2))
```
